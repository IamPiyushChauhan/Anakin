```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import time
```

```python
driver=webdriver.Chrome()
driver.get('https://food.grab.com/ph/en/')
driver.maximize_window()
print(driver.title)
```

```python
location = "Chong Boon Dental Surgery - Block 456 Ang Mo Kio Avenue 10, #01-1574, Singapore, 560456"
location_input =driver.find_element(By.ID,"location-input")
location_input.send_keys(location)
```

```python
search_btn = driver.find_element(By.XPATH,"//button[@class='ant-btn submitBtn___2roqB ant-btn-primary']")
search_btn.click()
```

```python
last_height = driver.execute_script("return document.body.scrollHeight")
new_height = 0

while True:
    driver.execute_script("window.scrollTo(0,document.body.scrollHeight)")
    time.sleep(10)
    new_height = driver.execute_script("return document.body.scrollHeight")
    if(new_height == last_height):
        break
    last_height = new_height
```
